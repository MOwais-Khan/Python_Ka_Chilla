# Adding Limits

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt
#load dataset
#sns.load_dataset("iris")  # sns se load krna he, iris pholo ka data set he #
phool= sns.load_dataset("iris") #aur jab data set load krlia to uska name b to dena he wo hamne phool dedia name
phool

#yaha tak ham run krke dekhlege k is dataset me kia he, dataset show krdege iris ka



#draw a line plot
#sns.lineplot(x="", y="", data=phool )     # yaha x-axis dena he y-axis dena he aur data jisko phool k equal rakhlia he


#x-axis me ham sepal_length lerahe he
#lineplot---- hamesha 2 numeric values k darmian banta he???
#alreasy discussed previus and will discuss more about it later in statistics chapter--- jinme numbers ayen khas tor pe cotinous variables hute he
# aur yeh catagorical he

sns.lineplot(x="sepal_length", y="sepal_width", data=phool ) 

#title hamne plt.show se pehle use krne hute he??
plt.title("Pholon ka plot")
#yaha x limit or y limit krni he
plt.xlim(2)  #(plt.x so show all related to x) #will do plt.xlim 
# on x axis graph starts from 4.5 till 8.0 now we watn from 2 ...
plt.ylim(1)   # on y axis graph starts from 2.5 now we watn from 1
#means based on your dataset you can set by your choice
plt.show()

#now we will be shown line plot  with title on  the top and with tuned x and y axis
```

# Styles

-whitegrid

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool= sns.load_dataset("iris") 
phool

sns.lineplot(x="sepal_length", y="sepal_width", data=phool ) 

plt.title("Pholon ka plot")
sns.set_style("darkgrid")   ## we used style-- dark
sns.set_style(style=None, rc=None)  # we used this because required to make none the default graph to use for the new style that we used below
plt.show()

#now we saw no changing just x-axis y-axis hataya he wo ... we added only sns.set_style("dark") but same graph so we all need to add
#grid
```

# Size of figure

```python
#import libraries
import seaborn as sns
import matplotlib.pyplot as plt

#load dataset
phool= sns.load_dataset("iris") 
phool

plt.figure(figsize=(2,3))
sns.lineplot(x="sepal_length", y="sepal_width", data=phool ) 

plt.title("Pholon ka plot")
sns.set_style("darkgrid")   ## we used style-- dark
sns.set_style(style=None, rc=None) 
plt.show()

```
